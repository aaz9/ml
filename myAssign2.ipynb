```python
Q1 = np.percentile(a['min_life_expectancy'], 25.) # 25th percentile of the data of the given feature
Q3 = np.percentile(a['min_life_expectancy'], 75.) # 75th percentile of the data of the given feature
IQR = Q3-Q1 #Interquartile Range
#outlier_step = IQR * 1.5 #That's we were talking about above
#outliers = feature_data[~((feature_data >= Q1 - outlier_step) & (feature_data <= Q3 + outlier_step))].index.tolist()
ll = Q1 - (1.5*IQR)
ul = Q3 + (1.5*IQR)
upper_outliers = a[a['min_life_expectancy'] > ul].index.tolist()
lower_outliers = a[a['min_life_expectancy'] < ll].index.tolist()
bad_indices = list(set(upper_outliers + lower_outliers))
drop = True
#if not drop:
    #print('For the feature {}, No of Outliers is {}'.format(temp['sepal.width'], len(bad_indices)))
if drop:
    a.drop(bad_indices, inplace = True, errors = 'ignore')
    print('Outliers from {} feature removed'.format(a['min_life_expectancy']))
```

```python
#Checking Outlier by definition and treating outliers

#getting median Age
Age_col_df = pd.DataFrame(a['min_life_expectancy'])
Age_median = Age_col_df.median()

#getting IQR of Age column
Q3 = Age_col_df.quantile(q=0.75)
Q1 = Age_col_df.quantile(q=0.25)
IQR = Q3-Q1

#Deriving boundaries of Outliers
IQR_LL = int(Q1 - 1.5*IQR)
IQR_UL = int(Q3 + 1.5*IQR)

#Finding and treating outliers - both lower and upper end
a.loc[a['min_life_expectancy']>IQR_UL , 'min_life_expectancy'] = int(Age_col_df.quantile(q=0.90))
a.loc[a['min_life_expectancy']<IQR_LL , 'min_life_expectancy'] = int(Age_col_df.quantile(q=0.01))
```

```python
#Separating features and response
'''features = ["Income","Age"]
response = ["Expense"]
x=MyData[features]
y=MyData[response]'''
aa = a
aa = aa.drop(['Name'] , axis = 1)
#print(x)

#Normalization/scaling of data - understanding scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(aa)
scaled_data

#converting data back to pandas dataframe
MyData_scaled = pd.DataFrame(scaled_data)
MyData_scaled.columns = aa.columns

#Separating features and response
features = ["trainability","energy","barking","playfulness","protectiveness","max_height_female","max_height_male","max_life_expectancy","min_life_expectancy"]
response = ["barking"]
x=MyData_scaled[features]
y=MyData_scaled[response]
```

```python
#Dividing data in test and train
#Importing neccesary packages
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=0)

#Fitting lineaar regression model
model = LinearRegression()
model.fit(x_train, y_train)

#Checking accuracy on test data
accuracy = model.score(x_test,y_test)
print(accuracy*100,'%')
```

```python
import numpy as np

y_pred = model.predict(x_test)

print('R2_score : ',metrics.r2_score(y_pred,y_test))
print('MSE : ',metrics.mean_squared_error(y_pred,y_test))
```
